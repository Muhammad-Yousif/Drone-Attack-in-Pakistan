```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# uploading datasets
df = pd.read_csv('drone-strikes-dataset.csv')
df.head()
```

## Datasets Shape (Rows & Columns)

## Dataset Summary

```python
df.shape
```

## Descriptive Statistics (description of data)

```python
df.describe()
```

## Sum of Peoples killed by each year


```python
# Number of Person kILLED BY EACH YEAR
plt.figure(figsize=(10,8))
sns.barplot(data=df['Killed'].groupby(by=df['Year']).sum())
plt.xlabel('Persons killed by Year')
plt.ylabel('sum of peopkles killed')
```

## AL-Qaeda fighters killed in each attack

```python
sns.barplot(data=df,y='Al-Qaeda',hue=df['No of Strikes'])
plt.xlabel('Number of strikes Occured on Al Qaeda')
plt.ylabel('Number of AL-Qaeda belongs killed in each attack')
```

## Maximum Civilians killed in each year

```python
plt.figure(figsize=(10,8))
sns.barplot(data=df,y=df['Civilians Max'].nlargest(13),hue='Year')
plt.xlabel('Maximum Civilians killed in each year')
```

## Number of Strikes on Taliban

```python
sns.barplot(data=df,y='Taliban',hue='No of Strikes')
plt.xlabel('Number of strikes Occured on Taliban')
plt.ylabel('Number of Talibans killed in each attack')
```

## Killed peoples belongs to

```python
summary_ = df[['Al-Qaeda','Taliban', 'Civilians Max', 'Civilians Min','Killed']].sum()
sns.barplot(data=summary_)
```

## Most hitted cities

```python
sns.barplot(data=df,y='No of Strikes',hue='City')
plt.xlabel('Most Hited Area')
```

## Total peoples killed belongs to diiferent groups

```python
sns.barplot(data=df[['Al-Qaeda','Taliban', 'Civilians Max', 'Civilians Min',]].sum())
plt.xlabel('Persons belong with')
plt.ylabel('Killed ')
```

## Number of strikes killed maximum peoples in each attack

```python
sns.barplot(data=df, x='No of Strikes',y='Killed')
plt.xlabel('Number of Strikes')
plt.ylabel('summation of persons Killed by number of strikes')
```

## Number of childs/womens killed 

```python
sns.barplot(data=df['Women/Children  '])
plt.xlabel('Womens and Childrens')
```

## Peoples killed belongs by each group

```python
data = df[['Civilians Max','Foreigners Max','Al-Qaeda','Taliban','Women/Children  ']]
sns.barplot(data=data)
plt.xlabel('killed Persons belongs')
```

## Maximum Foreigners killed in each city

```python
sns.barplot(data=df,y = 'Foreigners Max', hue='City')
plt.xlabel('Foreginers Hited at each city')
```

## Womens & Childrens Injured

```python
sns.barplot(data=df,y ='Injured ',hue='Women/Children  ')
plt.xlabel('Child and Womens Injured')
```
